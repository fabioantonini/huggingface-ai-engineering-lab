## Troubleshooting

| Problem | Solution |
|---|---|
| `ModuleNotFoundError` | Re-run the pip install cell above |
| Model download is slow | First run downloads model weights (~100MB–1GB); subsequent runs use cache |
| `CUDA out of memory` | Restart the kernel and use smaller `batch_size` or `max_length` |
| HF token required | Set `HUGGING_FACE_HUB_TOKEN` in `.env` at project root |
| Notebook not saving | Check Docker volume mount is active (`docker compose up`) |

> **Tip:** If you get stuck, restart the kernel (`Kernel → Restart & Clear Output`) and run cells from the top.
